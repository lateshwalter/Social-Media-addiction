```python
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=1 #20% of data will be used for testing and ensures reproducibility.
)
```

```python
# Initialize scaler
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
# Fit and transform the numerical data
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
def create_model(model):
    #train the model with 70% input and output
    model.fit(X_train,Y_train)
    #To test the model with 30% data, machine gives predicted output 
    Y_pred=model.predict(X_test)
    #create a report 
    print(confusion_matrix(Y_test,Y_pred))
    print(classification_report(Y_test,Y_pred))
    return model
```

```python
#Importing CONFUSION MATRIX and CLASSIFICATION REPORT class 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```
